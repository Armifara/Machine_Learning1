```python
df = df.drop_duplicates(keep="first").reset_index(drop=True)
```

# Step 3 - Separate X & Y (Weight)

```python
X = df.drop(columns=["id", "Weight"])
Y = df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

# Step 4 - Feature Selection, Make use of Pipelines
1. For Feature Selection, Ordinal Encoder is required
2. Ordinal Encoder - Categorical Columns

```python
# Separate categorical and numerical columns
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
# Convert categorical columns to category dtype
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="mean"), StandardScaler())
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder(),
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat),
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression

# Sequential Feature Selector is used to select features based on a model
```

```python
# Create a base model for feature selection
# Here we use Linear Regression as the base model
base_model = LinearRegression()

# Initialize the Sequential Feature Selector
# We will select 10 features using forward selection
# This is manually selecting features based on their importance
sel = SequentialFeatureSelector(
    base_model, n_features_to_select=10, direction="forward", n_jobs=-1
)

# n_jobs=-1 allows the use of all available CPU cores for parallel processing
# Fit the selector to the preprocessed data
```

```python
# To apply the feature selector, we need to fit it to the preprocessed data
sel.fit(X_pre, Y)

# Get the selected feature names
sel_cols = sel.get_feature_names_out()
```

```python
sel_cols
```

```python
# Check the number of selected features
len(sel_cols)
```

```python
# Keep n_features_to_select as "auto", if you want to automatically select features based on the model's performance,
sel = SequentialFeatureSelector(
    base_model, n_features_to_select="auto", direction="forward", n_jobs=-1
)
```

```python
sel.fit(X_pre, Y)

sel_cols = sel.get_feature_names_out()
```

```python
sel_cols
```

```python
len(sel_cols)
```

```python
# To select a particular column
sel_cols[0] # This will give you the first selected feature
```

```python
# To split the selected column, to separate the column name and the transformation applied
sel_cols[0].split("__")[
    1
]  # This will give you the column name and the transformation applied
# To get the original column name
# [1] will give you the column name after the transformation
```

```python
imp_cols = [col.split("__")[1] for col in sel_cols]
imp_cols
# This will give you the list of important columns after transformation
```

```python
X_imp = X[imp_cols]
```

```python
X_imp
# Now we can use the selected features for further analysis or modeling
```

### Feature Selection Complete

### For backward selection, just change the direction to backward from forward

# Step 5 - Create a Final Preprocessing Pipeline on Selected Features
1. Categorical - OneHotEncoding

```python
cat_sel = list(X_imp.columns[X_imp.dtypes == "object"])
con_sel = list(X_imp.columns[X_imp.dtypes != "object"])
# Convert categorical columns to category dtype
# What is dtype?
# # dtype is the data type of the column, it can be object, int, float, etc.
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

```python
cat_pipe2 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first"),
)
# handle_unknown="ignore" will ignore the unknown categories in the test set
# sparse_output=False will return a dense array instead of a sparse matrix
# drop="first" will drop the first category to avoid multicollinearity
```

```python
# To create the final preprocessing pipeline with the selected features
# IMP, to be applied on the selected features only
final_pre = ColumnTransformer(
    [
        ("num", num_pipe2, con_sel),
        ("cat", cat_pipe2, cat_sel),
    ]
).set_output(transform="pandas")
```

```python
X_imp_pre = final_pre.fit_transform(X_imp)
X_imp_pre.head()
```

# Step 6 - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X_imp_pre, Y, test_size=0.2, random_state=21
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

# Step 7 - Train the Model

```python
final_model = LinearRegression()
final_model.fit(xtrain, ytrain)
# Predicting on the test set
```

```python
final_model.intercept_
# This will give you the intercept of the model
```

```python
final_model.coef_
# Coefficients of the model
```

# Step 8 - Model Evaluation

```python
final_model.score(xtrain, ytrain)
```

```python
final_model.score(xtest, ytest)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    r2_score,
    root_mean_squared_error,
    mean_absolute_percentage_error,
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)

    mae = mean_absolute_error(y, ypred)
    rmse = root_mean_squared_error(y, ypred)
    r2 = r2_score(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)

    # Print the evaluation metrics
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAPE {mape:.2%}")
    print(f"R2: {r2:.2%}")
```

```python
print("Evaluation on Training Set:")
evaluate_model(final_model, xtrain, ytrain)
```

```python
print("Evaluation on Testing Set:")
evaluate_model(final_model, xtest, ytest)
```

### This Model is better with feature selection, R2 Score in Train and Test is >=0.8
