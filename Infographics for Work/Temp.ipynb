```python
import matplotlib.pyplot as plt

# Infographic 1: Bar chart comparing Revenue Growth, ASK Growth, Fleet Growth
metrics = ["Revenue Growth", "ASK Growth", "Fleet Growth"]
values = [49, 48, 50]

plt.figure()
plt.bar(metrics, values)
plt.title("FY25 Growth Metrics (%)")
plt.ylabel("Percentage (%)")
plt.tight_layout()
plt.show()

# Infographic 2: Timeline bar chart: Orders vs Delivered
years = ["2021", "2023", "2024"]
ordered = [72, 4, 150]
delivered = [30, 0, 0]

x = range(len(years))
width = 0.35

plt.figure()
plt.bar([i - width / 2 for i in x], ordered, width=width, label="Ordered")
plt.bar([i + width / 2 for i in x], delivered, width=width, label="Delivered")
plt.xticks(x, years)
plt.title("Boeing 737 MAX Orders vs Deliveries")
plt.xlabel("Year")
plt.ylabel("Number of Aircraft")
plt.legend()
plt.tight_layout()
plt.show()

# Infographic 3: Pie chart for International Operations Share
shares = [16, 84]  # 16% international, 84% domestic currently
labels = ["International", "Domestic"]

plt.figure()
plt.pie(shares, labels=labels, autopct="%1.0f%%", startangle=90)
plt.title("Current Operations: International vs Domestic")
plt.tight_layout()
plt.show()
```
